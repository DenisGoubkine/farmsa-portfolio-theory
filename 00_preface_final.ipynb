## The Point of This Project

Modern Portfolio Theory tells you **how to build an optimal portfolio** — minimize risk for a given return, or maximize the Sharpe ratio. The math is clean:

$$\mathbf{w}^* = \frac{\boldsymbol{\Sigma}^{-1}(\boldsymbol{\mu} - r_f\mathbf{1})}{\mathbf{1}^\top\boldsymbol{\Sigma}^{-1}(\boldsymbol{\mu} - r_f\mathbf{1})}$$

What MPT **doesn't tell you** is how to estimate $\boldsymbol{\Sigma}$ — the covariance matrix of asset returns. That's the open problem. That's what this project is about.

Each team member takes one estimator, derives it from scratch, implements it, and plugs it into the **same** portfolio engine. The final integration notebook runs all seven side by side. The question: does your choice of $\hat{\boldsymbol{\Sigma}}$ actually change the portfolio — and can you predict from theory when it should?

---
## What You Need to Know First

### The optimization problem

A portfolio $\mathbf{w}$ ($\mathbf{1}^\top\mathbf{w} = 1$, $\mathbf{w} \geq \mathbf{0}$) has expected return $\mu_p = \mathbf{w}^\top\boldsymbol{\mu}$ and variance $\sigma_p^2 = \mathbf{w}^\top\boldsymbol{\Sigma}\mathbf{w}$. We run two portfolio types:

- **MVO** — maximize Sharpe: $\max_\mathbf{w}\; (\mathbf{w}^\top\boldsymbol{\mu} - r_f)\,/\,\sqrt{\mathbf{w}^\top\boldsymbol{\Sigma}\mathbf{w}}$
- **Black-Litterman** — combine a market equilibrium prior $\boldsymbol{\Pi} = \delta\boldsymbol{\Sigma}\mathbf{w}_{\text{mkt}}$ with team views, then optimize on the posterior $\hat{\boldsymbol{\mu}}_{\text{BL}}$

In both cases, $\boldsymbol{\Sigma}$ is a direct input. Change the estimator, change the portfolio.

### Why the sample covariance isn't enough

The naïve estimator — $\hat{\boldsymbol{\Sigma}} = \frac{1}{T-1}\sum_t(\mathbf{r}_t - \bar{\mathbf{r}})(\mathbf{r}_t - \bar{\mathbf{r}})^\top$ — is unbiased but **inadmissible** (Stein, 1956): for $N \geq 3$, a better estimator always exists under mean squared error. In practice it distorts eigenvalues and produces concentrated, fragile portfolios. Every module is a different answer to this failure.

### What each module measures

Every notebook reports the same four metrics so results are directly comparable:

| Metric | Formula |
|:---|:---|
| Annualised return | $252 \times \bar{r}_p$ |
| Annualised volatility | $\sqrt{252} \times \text{std}(r_p)$ |
| Sharpe ratio | $(\hat{\mu}_p - r_f)\,/\,\hat{\sigma}_p$ |
| Concentration (HHI) | $\sum_i w_i^2$ — equal weight benchmark: $1/N$ |

## The Universe: Data Overview

Below we load and visualize the shared stock universe that all modules work from. This gives you the base statistical properties — what returns look like, how correlated assets are, and how the market has behaved over the sample period.

```python
# ---------------------------------------------------------
# Setup: Load stock universe
# ---------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from datetime import datetime

# Universe definition — adjust tickers and date range as needed
TICKERS = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'JPM', 'JNJ', 'XOM', 'PG', 'UNH', 'V', 
           'HD', 'MA', 'PFE', 'KO', 'PEP', 'MRK', 'ABBV', 'CVX', 'LLY', 'COST']
START_DATE = '2018-01-01'
END_DATE = '2024-12-31'

# Download adjusted close prices
print(f"Downloading {len(TICKERS)} stocks from {START_DATE} to {END_DATE}...")
prices = yf.download(TICKERS, start=START_DATE, end=END_DATE, auto_adjust=True)['Close']
prices = prices.dropna()

# Compute daily returns
returns = prices.pct_change().dropna()

print(f"\nUniverse loaded: {prices.shape[0]} trading days, {prices.shape[1]} assets")
print(f"Date range: {prices.index[0].strftime('%Y-%m-%d')} to {prices.index[-1].strftime('%Y-%m-%d')}")
```

```python
# ---------------------------------------------------------
# Plot 1: Normalized Price Trajectories
# ---------------------------------------------------------
normalized = prices / prices.iloc[0] * 100  # rebased to 100

fig, ax = plt.subplots(figsize=(12, 5))
for ticker in normalized.columns:
    ax.plot(normalized.index, normalized[ticker], alpha=0.7, linewidth=1, label=ticker)
ax.axhline(100, color='black', linestyle='--', linewidth=0.8, alpha=0.5)
ax.set_xlabel('Date')
ax.set_ylabel('Normalized Price (start = 100)')
ax.set_title('Price Trajectories of Universe Stocks')
ax.legend(loc='upper left', fontsize=7, ncol=4, framealpha=0.9)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------------
# Plot 2: Correlation Heatmap
# ---------------------------------------------------------
corr_matrix = returns.corr()

fig, ax = plt.subplots(figsize=(10, 8))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='RdYlBu_r', 
            center=0, vmin=-1, vmax=1, square=True, linewidths=0.5, 
            annot_kws={'size': 7}, ax=ax, cbar_kws={'shrink': 0.8})
ax.set_title('Pairwise Return Correlations')
plt.tight_layout()
plt.show()

# Summary stats
print(f"Correlation statistics:")
print(f"  Mean pairwise correlation: {corr_matrix.values[np.triu_indices_from(corr_matrix, k=1)].mean():.3f}")
print(f"  Min correlation: {corr_matrix.values[np.triu_indices_from(corr_matrix, k=1)].min():.3f}")
print(f"  Max correlation: {corr_matrix.values[np.triu_indices_from(corr_matrix, k=1)].max():.3f}")
```

```python
# ---------------------------------------------------------
# Plot 3: Return Distributions
# ---------------------------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: Histogram of all returns pooled
all_returns = returns.values.flatten()
axes[0].hist(all_returns, bins=100, density=True, alpha=0.7, color='steelblue', edgecolor='white')
axes[0].axvline(0, color='black', linestyle='--', linewidth=1)
axes[0].set_xlabel('Daily Return')
axes[0].set_ylabel('Density')
axes[0].set_title('Pooled Return Distribution')
axes[0].set_xlim(-0.15, 0.15)

# Right: Box plot by stock
returns_melted = returns.melt(var_name='Ticker', value_name='Return')
axes[1].boxplot([returns[t].values for t in returns.columns], labels=returns.columns, 
                patch_artist=True, boxprops=dict(facecolor='steelblue', alpha=0.6))
axes[1].axhline(0, color='black', linestyle='--', linewidth=0.8)
axes[1].set_xlabel('Ticker')
axes[1].set_ylabel('Daily Return')
axes[1].set_title('Return Distribution by Stock')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------------
# Summary Statistics Table
# ---------------------------------------------------------
stats = pd.DataFrame({
    'Ann. Return (%)': returns.mean() * 252 * 100,
    'Ann. Vol (%)': returns.std() * np.sqrt(252) * 100,
    'Sharpe': (returns.mean() * 252) / (returns.std() * np.sqrt(252)),
    'Skewness': returns.skew(),
    'Kurtosis': returns.kurtosis(),
    'Min (%)': returns.min() * 100,
    'Max (%)': returns.max() * 100
}).round(2)

stats = stats.sort_values('Sharpe', ascending=False)
print("Universe Summary Statistics (sorted by Sharpe):\n")
display(stats)
```

```python
# ---------------------------------------------------------
# Plot 4: Eigenvalue Spectrum of Sample Covariance
# ---------------------------------------------------------
cov_matrix = returns.cov()
eigenvalues = np.linalg.eigvalsh(cov_matrix)[::-1]  # sorted descending

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left: Eigenvalue spectrum
axes[0].bar(range(1, len(eigenvalues)+1), eigenvalues, color='steelblue', alpha=0.7)
axes[0].set_xlabel('Eigenvalue Index')
axes[0].set_ylabel('Eigenvalue')
axes[0].set_title('Sample Covariance Eigenvalue Spectrum')

# Right: Cumulative variance explained
cum_var = np.cumsum(eigenvalues) / np.sum(eigenvalues) * 100
axes[1].plot(range(1, len(eigenvalues)+1), cum_var, 'o-', color='steelblue', markersize=5)
axes[1].axhline(90, color='red', linestyle='--', linewidth=1, alpha=0.7, label='90% threshold')
axes[1].set_xlabel('Number of Components')
axes[1].set_ylabel('Cumulative Variance Explained (%)')
axes[1].set_title('Cumulative Variance Explained')
axes[1].legend()
axes[1].set_ylim(0, 105)

plt.tight_layout()
plt.show()

n_90 = np.argmax(cum_var >= 90) + 1
print(f"First {n_90} eigenvalues explain 90%+ of variance ({cum_var[n_90-1]:.1f}%)")
```
