## The Point

Modern Portfolio Theory gives you the *mechanics* of optimal portfolio construction. The tangency (max-Sharpe) portfolio has a closed-form solution:

$$\mathbf{w}^* = \frac{\boldsymbol{\Sigma}^{-1}(\boldsymbol{\mu} - r_f \mathbf{1})}{\mathbf{1}^\top \boldsymbol{\Sigma}^{-1}(\boldsymbol{\mu} - r_f \mathbf{1})}$$

The formula is settled. What is **not** settled is how you estimate $\boldsymbol{\Sigma}$ — the $N \times N$ covariance matrix of asset returns. That single input drives the entire portfolio. A distorted $\hat{\boldsymbol{\Sigma}}$ tilts weights, inflates leverage, and breaks diversification — even when the optimization is solved exactly.

**This project is about the estimator, not the optimizer.**

Each team member takes one estimator for $\boldsymbol{\Sigma}$, derives it from scratch, and plugs it into the shared portfolio engine. Everything else is held fixed. The final integration notebook runs all seven side by side and asks the question directly: *does your choice of $\hat{\boldsymbol{\Sigma}}$ change the portfolio — and when, and by how much?*


---

## Background

### Portfolio optimization

A long-only portfolio $\mathbf{w}$ satisfying $\mathbf{1}^\top\mathbf{w} = 1$, $\mathbf{w} \geq \mathbf{0}$ has:

$$\mu_p = \mathbf{w}^\top\boldsymbol{\mu}, \qquad \sigma_p^2 = \mathbf{w}^\top\boldsymbol{\Sigma}\,\mathbf{w}$$

We run two portfolio types that directly expose the role of $\boldsymbol{\Sigma}$:

| Portfolio | Objective |
|:---|:---|
| **MVO** (max-Sharpe) | $\displaystyle\max_{\mathbf{w}}\; \frac{\mathbf{w}^\top\boldsymbol{\mu} - r_f}{\sqrt{\mathbf{w}^\top\boldsymbol{\Sigma}\,\mathbf{w}}}$ |
| **Black-Litterman** | Blend market prior $\boldsymbol{\Pi} = \delta\boldsymbol{\Sigma}\mathbf{w}_{\text{mkt}}$ with team views; optimize on posterior $\hat{\boldsymbol{\mu}}_{\text{BL}}$ |

In both cases $\boldsymbol{\Sigma}$ is a direct input — change the estimator, change the portfolio.

### Why the sample covariance is not enough

The standard estimator is:

$$\hat{\boldsymbol{\Sigma}}_{\text{sample}} = \frac{1}{T-1}\sum_{t=1}^{T}(\mathbf{r}_t - \bar{\mathbf{r}})(\mathbf{r}_t - \bar{\mathbf{r}})^\top$$

It is unbiased — but **inadmissible** (Stein, 1956). For $N \geq 3$ a strictly better estimator exists under squared error loss. In practice, with $N$ close to $T$, the sample eigenvalues are severely distorted: large eigenvalues are inflated, small ones are deflated. The optimizer exploits these errors, producing concentrated and fragile portfolios. Every module in this project is a different answer to that failure.

### What each module reports

Every notebook outputs the same four metrics so results are directly comparable:

| Metric | Formula |
|:---|:---|
| Annualised return | $\hat{\mu}_p^{\text{ann}} = 252 \times \bar{r}_p$ |
| Annualised volatility | $\hat{\sigma}_p^{\text{ann}} = \sqrt{252}\;\text{std}(r_p)$ |
| Sharpe ratio | $S = (\hat{\mu}_p^{\text{ann}} - r_f)\;/\;\hat{\sigma}_p^{\text{ann}}$ |
| Concentration (HHI) | $H = \sum_i w_i^2$ &nbsp; (equal-weight benchmark: $1/N$) |

## The Universe

Below we load the shared stock universe — **50 equities** spanning five sectors — that all modules work from. Run this notebook once to generate the CSVs; every module notebook reads from those files directly.


```python
# ---------------------------------------------------------
# Setup: Load 50-stock universe
# ---------------------------------------------------------

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import yfinance as yf
import warnings
warnings.filterwarnings('ignore')

# ── Universe: 50 stocks across 5 sectors ──────────────────
TICKERS = [
    # Technology (15)
    'AAPL', 'MSFT', 'NVDA', 'GOOGL', 'META', 'AMZN', 'TSLA', 'ORCL',
    'INTC', 'AMD', 'ADBE', 'CRM', 'QCOM', 'TXN', 'AVGO',
    # Financials (10)
    'JPM', 'BAC', 'GS', 'MS', 'BLK', 'V', 'MA', 'AXP', 'C', 'WFC',
    # Healthcare (10)
    'JNJ', 'UNH', 'LLY', 'ABBV', 'MRK', 'PFE', 'ABT', 'TMO', 'AMGN', 'BMY',
    # Consumer (10)
    'PG', 'KO', 'PEP', 'WMT', 'HD', 'COST', 'MCD', 'SBUX', 'NKE', 'LOW',
    # Energy & Industrials (5)
    'XOM', 'CVX', 'COP', 'CAT', 'HON',
]

SECTORS = {
    **{t: 'Technology'         for t in ['AAPL','MSFT','NVDA','GOOGL','META','AMZN','TSLA','ORCL','INTC','AMD','ADBE','CRM','QCOM','TXN','AVGO']},
    **{t: 'Financials'         for t in ['JPM','BAC','GS','MS','BLK','V','MA','AXP','C','WFC']},
    **{t: 'Healthcare'         for t in ['JNJ','UNH','LLY','ABBV','MRK','PFE','ABT','TMO','AMGN','BMY']},
    **{t: 'Consumer'           for t in ['PG','KO','PEP','WMT','HD','COST','MCD','SBUX','NKE','LOW']},
    **{t: 'Energy & Industrials' for t in ['XOM','CVX','COP','CAT','HON']},
}

SECTOR_COLORS = {
    'Technology': '#2255a4',
    'Financials': '#c53030',
    'Healthcare': '#276749',
    'Consumer':   '#c05621',
    'Energy & Industrials': '#6b46c1',
}

START_DATE = '2018-01-01'
END_DATE   = '2024-12-31'

# ── Download ──────────────────────────────────────────────
print(f"Downloading {len(TICKERS)} stocks from {START_DATE} to {END_DATE}...")
prices = yf.download(TICKERS, start=START_DATE, end=END_DATE, auto_adjust=True)['Close']
prices = prices.dropna()
returns = prices.pct_change().dropna()

print(f"\nUniverse loaded: {prices.shape[0]} trading days × {prices.shape[1]} assets")
print(f"Date range : {prices.index[0].strftime('%Y-%m-%d')} → {prices.index[-1].strftime('%Y-%m-%d')}")
print(f"\nSector breakdown:")
for sector in dict.fromkeys(SECTORS.values()):
    tickers_in = [t for t in prices.columns if SECTORS.get(t) == sector]
    print(f"  {sector:<24} {len(tickers_in):>2} stocks")
```

```python
# ---------------------------------------------------------
# Plot 1: Normalized Price Trajectories (colour by sector)
# ---------------------------------------------------------

normalized = prices / prices.iloc[0] * 100

fig, ax = plt.subplots(figsize=(13, 5.5))

for ticker in normalized.columns:
    sector = SECTORS.get(ticker, 'Other')
    color  = SECTOR_COLORS.get(sector, '#888')
    ax.plot(normalized.index, normalized[ticker],
            color=color, alpha=0.55, linewidth=0.9)

# Sector legend proxies
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], color=SECTOR_COLORS[s], linewidth=2, label=s)
                   for s in SECTOR_COLORS]
ax.legend(handles=legend_elements, loc='upper left', fontsize=9, framealpha=0.9)

ax.axhline(100, color='black', linestyle='--', linewidth=0.8, alpha=0.4)
ax.set_xlabel('Date', fontsize=11)
ax.set_ylabel('Normalised Price  (start = 100)', fontsize=11)
ax.set_title('Price Trajectories — 50-Stock Universe (2018–2024)', fontsize=13, fontweight='bold')
ax.grid(True, alpha=0.25)
sns.despine(ax=ax)
plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------------
# Plot 2: Sector-Sorted Correlation Heatmap
# ---------------------------------------------------------

# Sort columns by sector so blocks are visible
sector_order = ['Technology','Financials','Healthcare','Consumer','Energy & Industrials']
sorted_tickers = [t for s in sector_order for t in prices.columns if SECTORS.get(t) == s]
corr = returns[sorted_tickers].corr()

fig, ax = plt.subplots(figsize=(13, 11))
sns.heatmap(corr, cmap='RdYlBu_r', center=0, vmin=-0.2, vmax=1,
            square=True, linewidths=0,
            cbar_kws={'shrink': 0.75, 'label': 'Pearson correlation'},
            ax=ax, xticklabels=True, yticklabels=True)

ax.tick_params(axis='x', labelsize=7, rotation=90)
ax.tick_params(axis='y', labelsize=7, rotation=0)
ax.set_title('Pairwise Return Correlations (sector-sorted)', fontsize=13, fontweight='bold', pad=14)

# Draw sector boundary lines
counts = [sum(1 for t in sorted_tickers if SECTORS.get(t) == s) for s in sector_order]
boundaries = [sum(counts[:i]) for i in range(1, len(counts))]
for b in boundaries:
    ax.axhline(b, color='white', linewidth=1.8)
    ax.axvline(b, color='white', linewidth=1.8)

plt.tight_layout()
plt.show()

off_diag = corr.values[np.triu_indices_from(corr, k=1)]
print(f"Mean pairwise correlation : {off_diag.mean():.3f}")
print(f"Median pairwise correlation: {np.median(off_diag):.3f}")
print(f"Min / Max               : {off_diag.min():.3f} / {off_diag.max():.3f}")
```

```python
# ---------------------------------------------------------
# Plot 3: Return Distributions
# ---------------------------------------------------------

fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))

# ── Left: pooled daily returns histogram ──────────────────
all_ret = returns.values.flatten()
axes[0].hist(all_ret, bins=120, density=True, color='#2255a4',
             alpha=0.75, edgecolor='none')
axes[0].axvline(0, color='black', linestyle='--', linewidth=1)
axes[0].set_xlabel('Daily Return', fontsize=11)
axes[0].set_ylabel('Density', fontsize=11)
axes[0].set_title('Pooled Return Distribution', fontsize=12, fontweight='bold')
axes[0].set_xlim(-0.15, 0.15)
axes[0].text(0.97, 0.96,
             f"kurt = {pd.Series(all_ret).kurtosis():.1f}",
             transform=axes[0].transAxes, ha='right', va='top',
             fontsize=10, color='#555')
sns.despine(ax=axes[0])

# ── Right: per-sector mean & vol scatter ─────────────────
ann_ret = returns.mean() * 252 * 100
ann_vol = returns.std() * np.sqrt(252) * 100

for sector, color in SECTOR_COLORS.items():
    tickers_s = [t for t in returns.columns if SECTORS.get(t) == sector]
    axes[1].scatter(ann_vol[tickers_s], ann_ret[tickers_s],
                    color=color, alpha=0.8, s=55, label=sector, zorder=3)
    for t in tickers_s:
        axes[1].annotate(t, (ann_vol[t], ann_ret[t]),
                         fontsize=6, alpha=0.7,
                         xytext=(3, 2), textcoords='offset points')

axes[1].axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.4)
axes[1].set_xlabel('Annualised Volatility (%)', fontsize=11)
axes[1].set_ylabel('Annualised Return (%)', fontsize=11)
axes[1].set_title('Risk–Return by Stock', fontsize=12, fontweight='bold')
axes[1].legend(fontsize=8, framealpha=0.9)
sns.despine(ax=axes[1])

plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------------------------
# Summary Statistics Table (50 stocks)
# ---------------------------------------------------------

stats = pd.DataFrame({
    'Sector'          : returns.columns.map(SECTORS),
    'Ann. Return (%)'  : returns.mean() * 252 * 100,
    'Ann. Vol (%)'     : returns.std() * np.sqrt(252) * 100,
    'Sharpe'           : (returns.mean() * 252) / (returns.std() * np.sqrt(252)),
    'Skewness'         : returns.skew(),
    'Excess Kurtosis'  : returns.kurtosis(),
    'Min Day (%)'      : returns.min() * 100,
    'Max Day (%)'      : returns.max() * 100,
}).round(2)

stats = stats.sort_values(['Sector', 'Sharpe'], ascending=[True, False])
print("Universe Summary Statistics (grouped by sector, sorted by Sharpe):\n")
display(stats)
```

```python
# ---------------------------------------------------------
# Plot 4: Eigenvalue Spectrum vs Marchenko-Pastur
# ---------------------------------------------------------

cov_matrix  = returns.cov()
eigenvalues = np.sort(np.linalg.eigvalsh(cov_matrix))[::-1]
T, N        = returns.shape
q           = T / N                         # aspect ratio

# Marchenko-Pastur upper / lower bounds
var_mean = np.mean(np.diag(cov_matrix))     # avg variance ≈ 1 for standardised data
lambda_plus  = var_mean * (1 + 1/np.sqrt(q))**2
lambda_minus = var_mean * (1 - 1/np.sqrt(q))**2

n_signal = np.sum(eigenvalues > lambda_plus)

fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))

# ── Left: eigenvalue bar chart with MP band ───────────────
axes[0].bar(range(1, N+1), eigenvalues, color='#2255a4', alpha=0.7, width=0.8)
axes[0].axhline(lambda_plus,  color='#c53030', linestyle='--', linewidth=1.5,
                label=f'MP upper bound ({lambda_plus:.4f})')
axes[0].axhline(lambda_minus, color='#276749', linestyle='--', linewidth=1.5,
                label=f'MP lower bound ({lambda_minus:.4f})')
axes[0].set_xlabel('Eigenvalue index', fontsize=11)
axes[0].set_ylabel('Eigenvalue', fontsize=11)
axes[0].set_title('Sample Covariance Eigenvalue Spectrum', fontsize=12, fontweight='bold')
axes[0].legend(fontsize=9)
axes[0].set_xlim(0, N+1)
sns.despine(ax=axes[0])

# ── Right: cumulative variance explained ─────────────────
cum_var = np.cumsum(eigenvalues) / np.sum(eigenvalues) * 100
axes[1].plot(range(1, N+1), cum_var, 'o-', color='#2255a4', markersize=4, linewidth=1.5)
axes[1].axhline(90, color='#c53030', linestyle='--', linewidth=1.2,
                alpha=0.8, label='90 % threshold')
axes[1].fill_between(range(1, N+1), cum_var, alpha=0.12, color='#2255a4')
axes[1].set_xlabel('Number of components', fontsize=11)
axes[1].set_ylabel('Cumulative variance explained (%)', fontsize=11)
axes[1].set_title('Cumulative Variance Explained', fontsize=12, fontweight='bold')
axes[1].legend(fontsize=9)
axes[1].set_ylim(0, 105)
axes[1].set_xlim(0, N+1)
sns.despine(ax=axes[1])

plt.tight_layout()
plt.show()

n_90 = np.argmax(cum_var >= 90) + 1
print(f"Aspect ratio q = T/N  : {q:.2f}")
print(f"Marchenko-Pastur upper: {lambda_plus:.5f}")
print(f"Signal eigenvalues (> MP upper) : {n_signal}  out of {N}")
print(f"Components needed for 90%% variance: {n_90}")
```

**What to notice in the data:**

- **Block structure in the heatmap** — correlations are clearly higher within sectors than across them. This motivates factor models (M3, M4) and constant-correlation (M7).
- **Concentrated eigenvalue spectrum** — a handful of eigenvalues dominate. The Marchenko-Pastur bound separates the signal (market/sector factors) from the noise component. This is what RMT (M6) explicitly targets.
- **Fat tails** — excess kurtosis is large for most stocks. The Normal assumption in MPT is an approximation; DCC-GARCH (M5) can adapt to it.
- **Wide return dispersion** — individual Sharpe ratios range from negative to above 2, so the estimator can meaningfully tilt weights.

This is the raw material. Every module takes this exact dataset and applies a different estimator to extract $\hat{\boldsymbol{\Sigma}}$ for portfolio optimization.
